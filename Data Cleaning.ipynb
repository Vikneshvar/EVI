```python
df_age= df_age.rename(columns={'TractID': 'TRACT'})
df_age
```

```python
df_final=df_tract.merge(df_age,on='TRACT',how='inner')
```

```python
df_final
```

```python
df_final.sort_values(by=['Geography','ZIP'],inplace=True)
```

```python
df_final
```

```python
df_final.drop(labels=['TRACT','RES_RATIO','BUS_RATIO','OTH_RATIO','TOT_RATIO'],axis=1,inplace=True)
```

```python
arr_variables=df_final.columns 
arr_variables.unique()
```

```python
variables=[]
for ss in arr_variables:
    if 'labor force' in ss:
        pass
    else:
        variables.append(ss)
```

```python
variables
```

```python
retain_variables=[]
for each in variables:
    if('ZIP' ==each or 'Geography' ==each or 'Estimate; Total:'== each or 'Estimate; Male:'==each or 'Estimate; Female:'== each or '65 to 69 years:' in each or '70 to 74 years:' in each or '75 years' in each):
        retain_variables.append(each)
    else:
        pass 
```

```python
retain_variables
```

```python
df_final=df_final[retain_variables]
```

```python
df_final
```

```python
df_final[' Male: Age: 65 and over']=df_final[df_final.columns[4]]+df_final[df_final.columns[5]]+df_final[df_final.columns[6]]
df_final[' Female: Age: 65 and over']=df_final[df_final.columns[8]]+df_final[df_final.columns[9]]+df_final[df_final.columns[10]]
df_final[' Overall People Age: 65 and over']=df_final[df_final.columns[11]]+df_final[df_final.columns[12]]
```

```python
df_final
```

```python
df_orange=df_final[df_final['Geography'].str.contains('Orange')]
```

```python
df_la=df_final[df_final['Geography'].str.contains('Los')]
```
